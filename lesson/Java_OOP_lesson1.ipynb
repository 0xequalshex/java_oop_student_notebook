# **What is OOP?**

**Object-Oriented Programming (OOP)** is a methodology for building software by thinking about it in terms of objects

Instead of writing one massive continuous program we break the logic into smaller and self-contained pieces, and this pieces in code called *objects*

object exists in memory and has two primary roles:

**- Fields**: for Storing data ( like **int x;** )

**- Methods**: for Performing operations ( like **public void calc(double a)** )

# **What is the difference between Objects and Classes?**

To understand the difference think of a blueprint and house


**- Class**: This is the code "blueprint". It describes what the object will be, what data it will hold, and what it will do

**- Object**: This is the actual "instance" built from that blueprint. You can use one class to create as many objects as you need in memory

as an exmple the **Scanner**

```java
import java.util.Scanner; //this is the API, it's a way to use classes other people made

// now we need to make a copy of this class, and this is the object

Scanner keyboard = new Scanner(System.in);

// now we have a copy of class in our memory and you can use the methods that inside the class
System.out.print("write number: ");
int meow = keyboard.nextInt(); // keyboard is the object or copy of class, nextInt is a method inside the object

System.out.print("the number you put is: " + meow);
```

```java
// now must of the work will be inside the class we will work on the main later

class Rectangle {
    // we need now to make a Fields to work with

    // the private word it mean the Field can only be accessed by methods within the same class
    // and there is public too who can be accessed by code inside or outside the class
    private double length;
    private double width;

    // after we make a Dields we will also make a methods
    
    // in method we need two type of main methods that programmers use:
    
    // Mutator: Method that modify the field

    public void setLength(double len){
        length = len;
    }

    // same thing for other field
    public void setWidth(double wid){
        width = wid;
    }

    // now after we modify the field we need to make class give the main this values again
    
    // but wait why we need to give the main this value isn't we the programmers who give the class this value from the begging???!
    // actully we give him the value but not save it in the main
    // in real life situation for security it's a bad thing to make the values in the main we need to hide it from others to not modify it easly
    // not just that, for make the class easy to copy we need this to put a values more easly

    // now lets continue the code

    
    // Accessors is the method we use to return the value to the main

    public double getLength(){
        return length;
    }

    // same thing for other field
    public double getWidth(){
        return width;
    }

    // now we really close to the main goal of our class, calculate the area and this easy just make a methods that return the area value

    public double getArea(){
        return length * width;
    }
    
    // why we dont just make a field for area?? this will be explained in next lesson
}
```

**congratulations** you made your first class!!

now we will make the first copy of our class

```java
// lets start use our main class

class ProjactName {
    public static void main(String[] args) {
        // first of all you can't use your class without make a copy so lets just make it

        Rectangle box = new Rectangle();

        // now we will use our methods
        box.setLength(5);
        box.setWidth(3);
        

        // print them
        System.out.println("The box length is: " + box.getLength());
        System.out.println("The box width is: " + box.getWidth());
        System.out.println("The box area is: " + box.getArea());
        
    }
}
```

# **full code**

```java
class Rectangle {
    
    private double length;
    private double width;

    public void setLength(double len){
        length = len;
    }

    public void setWidth(double wid){
        width = wid;
    }

    public double getLength(){
        return length;
    }

    public double getWidth(){
        return width;
    }

    public double getArea(){
        return length * width;
    }
    
}


class ProjactName {
    public static void main(String[] args) {

        Rectangle box = new Rectangle();

        box.setLength(5);
        box.setWidth(3);
        
        System.out.println("The box length is: " + box.getLength());
        System.out.println("The box width is: " + box.getWidth());
        System.out.println("The box area is: " + box.getArea());
        
    }
}
```
